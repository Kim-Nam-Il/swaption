# BDT Tree

```python
sigma = 0.2
sigma_sqrt_dt = sigma*np.sqrt(1/12)
sigma_sqrt_dt_lst = [sigma_sqrt_dt] * 119
sigma_array = np.array(sigma_sqrt_dt_lst)
r0 = 0.0174
y = np.log(r0)
theta = 0.025674
x = np.linspace(theta-1,theta+1, 119)
theta_array = np.array([0]+ list(x))
theta_array = i3 * theta_array

sigma_array = np.concatenate([np.array([0]) , sigma_array])
```

```python

def make_BDT_node(r0, theta_array, sigma_array) : 
    global  i , i2, i3 , node_diff
    y = np.log(r0)
    cumsum_sigma_array = sigma_array.cumsum()
    cummulated_sigma = i2 * cumsum_sigma_array
    diff = sigma_array * node_diff
    diffusion_mat = cummulated_sigma + diff
    init_mat = y*i2
    y_plus_theta = init_mat + theta_array * 1/12
    y_plus_theta_sigmadt = y_plus_theta + diffusion_mat
    return np.exp(y_plus_theta_sigmadt) - np.rot90(np.rot90(i3))
```

```python
def Swaption_price(S_a_b, Annuity,sigma, K, T, PV_0_T) :
    d1 = (np.log(S_a_b/K) + (0.5* sigma**2)*T  )/(sigma * np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)
    Nd1 = norm.cdf(d1)
    Nd2 = norm.cdf(d2)
    
    P = N * Annuity * PV_0_T * ( S_a_b * Nd1 - K * Nd2)
    return P
```
